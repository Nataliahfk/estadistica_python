## Importando la biblioteca statsmodels

https://www.statsmodels.org/stable/index.html

```python
import statsmodels.api as sm
```

## Estimando el modelo con statsmodels

```python
X_train_con_contante=sm.add_constant(X_train)
X_train_con_contante
```

```python
# sm.OLS minimos cuadrados ordinarios para lOLS es la traducción de mínimos cuadrados ordinarios.
# es el método que estima el modelo de regresión lineal.
# Esta función nos pide con parámetros de entrada la variable dependiente y, pero recuerden que
# como estamos trabajando con datos de entrenamiento para estimar el modelo será y_train.
# Nos pide el DataFrame de las variables explicativas con la constante incluida, que hemos creado aquí arriba, es X_train_con_constante.
# parámetro que se llama hasconst=. Esta función es necesaria pues necesitamos decirle a los statsmodels que la constante fue añadida al DataFrame
# de las variables explicativas
modelo1=sm.OLS(y_train,X_train_con_contante, hasconst=True).fit()
```

## Instanciando la clase *LinearRegression()*

```python
# Crear el modelo
modelo= LinearRegression()
```

## Utilizando el método *fit()* del objeto "modelo" para estimar nuestro modelo lineal con los datos de ENTRENAMIENTO (y_train y X_train)

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.fit

📌 Qué hace exactamente<br>

Recibe los datos de entrada (X_train) y las etiquetas o valores objetivo (y_train).

- X_train → matriz de características (variables explicativas).
- y_train → vector o serie con la variable que quieres predecir.

Guarda el modelo entrenado en el objeto modelo

Después podrás usar .predict(...) para hacer predicciones con datos nuevos.

Si quieres confirmar que el entrenamiento se hizo, puedes inspeccionar atributos del modelo (por ejemplo, en regresión lineal modelo.coef_ y modelo.intercept_).

```python
# entrenamiento de un modelo de machine learning
modelo.fit(X_train,y_train)
print(modelo.coef_)
print(modelo.intercept_)
```

## Obteniendo el coeficiente de determinación (R²) del modelo estimado con los datos de ENTRENAMIENTO


https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.score

### Coeficiente de Determinación - R²

El coeficiente de determinación (R²) es una medida de resumen que indica cuánto se ajusta la recta de regresión a los datos. Es un valor entre 0 y 1.


$$R^2(y, \hat{y}) = 1 - \frac {\sum_{i=0}^{n-1}(y_i-\hat{y}_i)^2}{\sum_{i=0}^{n-1}(y_i-\bar{y}_i)^2}$$

```python
# devuelve el coeficiente de determinación R² para los datos de entrenamiento.
print('R²: {}'.format(round(modelo.score(X_train,y_train),(3))))
```

## Generando predicciones para los datos de PRUEBA (X_test) con el método *predict()* del objeto "modelo"

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.predict

hace que el modelo ya entrenado (modelo) genere predicciones para los datos de prueba X_test.

📌 Paso a paso
- Usa los parámetros aprendidos en .fit()
- Procesa cada fila de X_test para estimar el valor de salida correspondiente.
- Devuelve un array/serie con las predicciones → en y_pred.



```python
y_pred = modelo.predict(X_test)

# para revisar que este correctamente generado
print(modelo.coef_)
print(modelo.intercept_)
```

## Obteniendo el coeficiente de determinación (R²) para las predicciones del nuestro modelo

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score

```python
# coeficiente de determinación R² pero comparando las predicciones (y_pred) con los valores reales (y_test).
print('R²: {}'.format(round(metrics.r2_score(y_test,y_pred),(3))))

# Valor muy similar al R² del conjunto entrenamiento / el modelo es capaz de producir predicciones a los valores reales.
```

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>Un aspecto interesante del modelo log-lineal, que lo ha hecho ampliamente utilizado en estudios aplicados, es que los coeficientes angulares $\beta_2$ y $\beta_3$ miden las elasticidades de Y en relación a $X_2$ y $X_3$, es decir, la variación porcentual de Y correspondiente a una variación porcentual dada (pequeña ) en $X_2$ y $X_3$.</p>

La pendiente (coef_) indica cuánto aumenta y por cada unidad que aumenta x.

## Gráfico de dispersión entre valor estimado y valor real

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
ax = sns.scatterplot(x= y_predictivo_entrenamiento, y = y_train)
ax.figure.set_size_inches(12, 6)
ax.set_title('Predicción X Real', fontsize=18)
ax.set_xlabel('log del Precio - Predicción', fontsize=14)
ax.set_ylabel('log del Precio - Real', fontsize=14)

plt.show()
```

## Obteniendo los residuos

```python
# diferencia entre los valores reales y los valores predecidos
residuo = y_train - y_predictivo_entrenamiento
```

## Graficando la distribuición de frecuencias de los residuos

```python
ax = sns.histplot(residuo, kde=True)
ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuición de Frecuencias de los Residuos', fontsize=18)
ax.set_xlabel('log del Precio', fontsize=14)

plt.show()
```
